Data Model with Gradient Boosting

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection

#split data
test=['Make_Code', 'Model_Code', 'Year', 'SAE Net Horsepower @ RPM', 'Gas Mileage']
x_variables = ds[test] #columns tested on
y_variables = ds['MSRP'] #tested column
x_training, x_testing, y_training, y_testing = model_selection.train_test_split(x_variables, y_variables, random_state=42)
#score result 75% goes to training 25% is what is actually tested.
```

```python
gb = GradientBoostingRegressor()
gb.fit(x_training, y_training) #fit training data 
```
